```python
import pandas as pd

data = pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")

data = data.dropna(subset=['BOROUGH'])

# Group by borough and count crashes
crashes_by_borough = data['BOROUGH'].value_counts()

# Find the borough with the highest total number of crashes
highest_crashes_borough = crashes_by_borough.idxmax()
highest_crashes_count = crashes_by_borough.max()

print(f"The borough with the highest total number of crashes since 2012 is {highest_crashes_borough} with {highest_crashes_count} crashes.")
```

```python
import pandas as pd

# Define the population estimates
population_data = {
    'Borough': ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island'],
    'Population': [1446788, 2648452, 1638281, 2330295, 487155]
}

population_df = pd.DataFrame(population_data)

crash_data = pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")

# Filter data for records since July 2012
crash_data['CRASH DATE_CRASH TIME'] = pd.to_datetime(crash_data['CRASH DATE_CRASH TIME'])
crash_data = crash_data[crash_data['CRASH DATE_CRASH TIME'] >= '2012-07-01']

# Calculate the total number of crashes for each borough during this period
total_crashes_by_borough = crash_data.groupby('BOROUGH')['NUMBER OF PERSONS INJURED'].sum().reset_index()

# Calculate crashes per 100,000 people without using merge
result_data = pd.merge(total_crashes_by_borough, population_df, left_on='BOROUGH', right_on='Borough')
result_data['Crashes per 100,000 People'] = (result_data['NUMBER OF PERSONS INJURED'] / result_data['Population']) * 100000

# Check if there is any data available
if not result_data.empty:
    # Find the borough with the highest number of crashes per 100,000 people
    max_crashes_borough = result_data.loc[result_data['Crashes per 100,000 People'].idxmax()]
    print(f"The borough with the most crashes for every 100,000 people is {max_crashes_borough['BOROUGH']} with {max_crashes_borough['Crashes per 100,000 People']:.2f} crashes per 100,000 people.")
else:
    print("Cannot be calculated with the given information")
```

```python
import pandas as pd

crash_data = pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")

# Filter out rows where either 'VEHICLE TYPE CODE 1' or 'VEHICLE TYPE CODE 2' is missing
crash_data = crash_data.dropna(subset=['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2'])

# Group by combinations of vehicle types and count the accidents
vehicle_combinations = crash_data.groupby(['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']).size().reset_index(name='Accident Count')

# Find the combination with the most accidents
most_common_combination = vehicle_combinations.loc[vehicle_combinations['Accident Count'].idxmax()]

print(f"The combination of vehicles with the most accidents is: {most_common_combination['VEHICLE TYPE CODE 1']} and {most_common_combination['VEHICLE TYPE CODE 2']} with {most_common_combination['Accident Count']} accidents.")
```

```python
import pandas as pd

crash_data = pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")

# Group by contributing factors and count the accidents
crash_data = crash_data[crash_data['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']

# Group by contributing factors and count the accidents
top_factors = crash_data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().reset_index()
top_factors.columns = ['Contributing Factor', 'Count']

# Select the top 3 contributing factors
top_3_factors = top_factors.head(3)

# Calculate the total number of accidents
total_accidents = len(crash_data)

# Calculate the total count of accidents for the top 3 contributing factors
total_top_3_accidents = top_3_factors['Count'].sum()

# Calculate the proportion of total crashes accounted for by the top 3 contributing factors
proportion_top_3 = total_top_3_accidents / total_accidents

print(f"The top 3 causes of crashes (ignoring 'Unspecified') account for approximately {proportion_top_3:.2%} of total crashes.")
```

```python
import pandas as pd

crash_data = pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")

# Filter out rows with 'CONTRIBUTING FACTOR VEHICLE 1' as 'Unspecified'
crash_data = crash_data[crash_data['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']

# Filter out rows where 'CONTRIBUTING FACTOR VEHICLE 1' is known
known_cause_crashes = crash_data[crash_data['CONTRIBUTING FACTOR VEHICLE 1'].notna()]

# Group by contributing factors and count the accidents
leading_cause = known_cause_crashes['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().reset_index()
leading_cause.columns = ['Contributing Factor', 'Count']

# Select the leading cause (the one with the highest count)
leading_cause = leading_cause.head(1)

# Calculate the total number of accidents
total_accidents = len(known_cause_crashes)

# Calculate the proportion of accidents attributable to the leading cause
proportion_leading_cause = leading_cause['Count'].iloc[0] / total_accidents

print(f"The leading cause of crashes (ignoring 'Unspecified') is '{leading_cause['Contributing Factor'].iloc[0]}' with a proportion of approximately {proportion_leading_cause:.2%} of accidents.")
```
